```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
missing = pd.read_csv('missing_gene_test_results.csv')
for i in range(len(missing)):
    if(int(missing.iloc[i]['Percent Missing']) != int(missing.iloc[i]['Missing Data'])):
        print(missing.iloc[i]['Percent Missing'], int(missing.iloc[i]['Missing Data']))
```

```python
sns.lineplot(x='Percent Missing', y='Missing Data', data=missing)
plt.title('Missing Data vs Percent Missing Genotypes (100 Genotypes Total)')
plt.xlabel('Percent Missing Input')
plt.ylabel('Actual Missing Genotypes')
plt.show()
```
