```python
import pandas as pd
import numpy as np
```

```python
pixy_df1 = pd.read_csv('PixyVCFSim.csv', sep = '\t')
vcftools_df2 = pd.read_csv('VCFToolsVCFSim.csv', sep = '\t')
```

```python
pixy_df1
```

```python
vcftools_df2
```

```python
pixy_df1 = pixy_df1.round({"vcfsim_pi":3})
pixy_frequencies = pixy_df1.groupby(['vcfsim_pi']).size().reset_index(name = 'count')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd


sb.lineplot(data=pixy_frequencies, x='vcfsim_pi', y='count')
plt.xlabel('VCFSim Pi')
plt.ylabel('Frequency')
plt.title('Line Plot of Pi Values with Frequency')
plt.axvline(x=4 * 1700000 * 1e-9, color='red', linestyle='--', linewidth=2)
plt.axvline(x=pixy_df1['vcfsim_pi'].mean(), color='blue', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()
```

```python
vcftools_df2 = vcftools_df2.round({"vcfsim_pi":3})
vcftools_frequencies = vcftools_df2.groupby(['vcfsim_pi']).size().reset_index(name = 'count')
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
import pandas as pd


sb.lineplot(data=vcftools_frequencies, x='vcfsim_pi', y='count')
plt.xlabel('VCFSim Pi')
plt.ylabel('Frequency')
plt.title('Line Plot of Pi Values with Frequency')
plt.axvline(x=4 * 1700000 * 1e-9, color='red', linestyle='--', linewidth=2)
plt.axvline(x=vcftools_df2['vcfsim_pi'].mean(), color='blue', linestyle='--', linewidth=2)

plt.tight_layout()
plt.show()
```
